```python
import pandas as pd
import numpy as np

def gradient_descent(x, y, learning_rate=0.001, iterations=100):
    m_curr = b_curr = 0  # IniÈ›ializare coeficienti
    n = len(x)
    
    for i in range(iterations):
        
        y_predicted = m_curr * x + b_curr
        
        
        cost = (1/n) * sum([val ** 2 for val in (y - y_predicted)])
        
     
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
  
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
     
        if i % 10 == 0: 
            print(f"m {m_curr}, b {b_curr}, iteration {i}, cost {cost}")
    
    return m_curr, b_curr

def predict_future_fire_area_per_country(data_path, future_years, learning_rate=0.001, iterations=100):
   
    data = pd.read_csv(data_path)
    

    predictions = {}
    
 
    grouped_data = data.groupby('Entity')
    
    for country, group in grouped_data:
        x = group['Year'].values
        y = group['area burnt by wildfires in 2024'].values
        
   
        m, b = gradient_descent(x, y, learning_rate, iterations)
        
    
        future_years_array = np.array(future_years)
        future_area_predictions = m * future_years_array + b
        

        predictions[country] = dict(zip(future_years, future_area_predictions))
    
    return predictions


data_path = 'areaburntbywildfiresbyweek new.csv'  
future_years = [2025, 2026, 2027]  
predictions = predict_future_fire_area_per_country(data_path, future_years)
for country, prediction in predictions.items():
    print(f"\nPrediction for {country}:")
    for year, area in prediction.items():
        print(f"Year {year}: {area:.2f} kmÂ²")
```
